```python
import pandas as pd
import numpy as np
import gensim
```

```python
dataset = pd.read_csv('dataset.csv')
dataset.reset_index(inplace=True,drop=True)
dataset
```

```python
model = gensim.models.Word2Vec.load("word2vec.model")
```

```python
dataset['embed'] = ''
dataset['embed'] = dataset['embed'].astype('object')

for index,row in dataset.iterrows():
    text = row.cleaned_text.split()
    embed = np.zeros(100)
    for word in text:
        try:
            embed = embed + np.array(model.wv[word]).reshape(100)
        except:
            pass
    embed = embed/len(text)
    dataset['embed'][index] = embed.tolist()
    if (index%1000 == 0):
        print(index)
```

```python
dataset.to_csv('embeds.csv')
```
